```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. Data Loading
file_path = "/content/car_rental_data.csv"  # Modify if your file path is different
df = pd.read_csv(file_path)
```

```python
print(df.head())  # Display first few rows
print(df.info())  # Get data types and missing values
print(df.describe())  # Summary statistics for numerical columns
```

```python
# 3. Data Preprocessing (If needed)
# Example: Extracting brand and model from 'desc_vehicle'
df[['brand', 'model']] = df['desc_vehicle'].str.split('-', n=1, expand=True)
```

```python
# a. Segment Distribution
plt.figure(figsize=(8, 6))
df['segment'].value_counts().plot(kind='bar', color=['skyblue', 'lightcoral', 'lightgreen', 'gold'])
plt.title('Market Segmentation - Distribution of Customers')
plt.xlabel('Segment')
plt.ylabel('Number of Customers')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# b. Purchase Value by Segment
plt.figure(figsize=(10, 6))
sns.boxplot(x='segment', y='purchase_value', data=df, palette="Set2")
plt.title('Purchase Value Distribution across Segments')
plt.xlabel('Segment')
plt.ylabel('Purchase Value')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# c. Contract Duration by Segment
plt.figure(figsize=(10, 6))
sns.violinplot(x='segment', y='contract_duration', data=df, palette="pastel")
plt.title('Contract Duration Distribution across Segments')
plt.xlabel('Segment')
plt.ylabel('Contract Duration')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# d. Brand Preference by Segment (If you have a 'brand' column)
plt.figure(figsize=(12, 6))
brand_segment_counts = df.groupby(['segment', 'brand'])['brand'].count().unstack().fillna(0)
brand_segment_counts.plot(kind='bar', stacked=True, colormap='viridis')
plt.title('Brand Preference across Segments')
plt.xlabel('Segment')
plt.ylabel('Number of Customers')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Brand', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
